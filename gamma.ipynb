```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

tokenizer = AutoTokenizer.from_pretrained("google/gemma-2b-it")
model_gemma_2b_it = AutoModelForCausalLM.from_pretrained(
    "google/gemma-2b-it",
    torch_dtype=torch.bfloat16
)
```

```python
input_text = "Write me a poem about Machine Learning."
input_ids = tokenizer(input_text, return_tensors="pt")

outputs = model_gemma_2b_it.generate(**input_ids)
print(tokenizer.decode(outputs[0]))
```

#### 1. AI Detection

```python
instruction_prompt = (
    "You are a content detection expert.\n"
    "Your task is to determine whether the following student-written paragraph is generated by an AI or a human.\n"
    "Be honest, unbiased, and give a reason for your decision.\n\n"
    "Student Paragraph:\n"
    "\"\"\"\n"
    "{}\n"
    "\"\"\"\n\n"
    "Your Analysis:"
)
```

```python
def detect_ai_content_instructional(student_paragraph):
    # Format the full prompt with instructions
    prompt = instruction_prompt.format(student_paragraph)

    # Tokenize and generate
    inputs = tokenizer(prompt, return_tensors="pt").to(model_gemma_2b_it.device)
    outputs = model_gemma_2b_it.generate(**inputs, max_new_tokens=250)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return response
```

```python
student_input = "Machine learning allows systems to learn from data and make predictions without being explicitly programmed."

result = detect_ai_content_instructional(student_input)
print(result)
```

#### 2. Feedback Generation

```python
feedback_instruction_prompt = (
    "You are a subject matter expert reviewing a student's writing submission.\n"
    "Your task is to provide constructive feedback based on the following:\n"
    "- Does the paragraph fully address the topic?\n"
    "- Are any key concepts or important points missing?\n"
    "- Is the information complete and accurate?\n"
    "- Is the writing logically structured and easy to follow?\n"
    "- Offer clear suggestions for improvement if needed.\n\n"
    "- Writing follows the completeness and also structure of report/ research/ concise etc check according to that "
    "Student Paragraph:\n"
    "\"\"\"\n"
    "{}\n"
    "\"\"\"\n\n"
    "Provide helpful, respectful feedback to guide the student:"
)
```

```python
def generate_student_feedback(student_paragraph):
    # Format the full prompt with feedback instructions
    prompt = feedback_instruction_prompt.format(student_paragraph)

    # Tokenize and generate
    inputs = tokenizer(prompt, return_tensors="pt").to(model_gemma_2b_it.device)
    outputs = model_gemma_2b_it.generate(**inputs, max_new_tokens=300)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return response
```

```python
student_assignment = "Executive Summary This report presents the results of a customer data analysis to better understand the spending patterns, online shopping habits, and uses of discount within AussieMart’s customer base. The data used for this analysis has 400 randomly selected customers from various demographics. With the analysis, we found that spending level of customer vary widely, with an average of $49,000 spending per customer annually. Most of the customer tend to do online shopping and, higher percentage of customers saying that, they shop online regularly. There is asumption that, when there will be more transaction then the more spending will be, to carry this we have done analysis to find the link between them spending and trasaction. The loyalty program participation shows higher males compared to females in online shoping. Furthermore, the discount usage is influenced by age as younger customers more likely to use discounts frequently. From the analysis this has been observed that average product return rate is higher than the market. Eventually, to improve the accuracy in future results, we recommend to increase the number of survey participants to at least 536 customers. This would allow for better estimation of average spending and online shopping habits. Introduction AussieMart is one of the Australia’s most famous retail brands, which is known for offering a wide range of products through both physical stores and an online platform. To initiate better marketing and customer engagement strategies, this report analyze and examine customer’s behaviour using a sample dataset of 400 randomly chosen customers. This report covers five main areas: •	Understanding customer behaviour by analyzing spending and online shopping frequency. •	Exploring whether there is any relationships between customer characteristics and behaviour. •	Estimating key averages and proportions based on sample data. •	Testing external claims about the retail market. •	And determining the sample size which is needed for future surveys to make better and reliable decisions."
result = generate_student_feedback(student_assignment)
print(result)
```

## Model Versions & Historical Context
### 1. Gemma 2B-it

- The original instruction‑tuned variant in the first Gemma generation ("Gemma 1").

- Contains ~2 billion parameters, geared for efficiency on constrained hardware.

- Effective lightweight model, suitable for apps like question‑answering and summarization.

### 2. Gemma‑2‑2B‑it

- A newer Gemma 2 model variant with ~2.6 billion parameters.

- Represents a successor-generation version of Gemma 2B with enhanced architecture.

| Factor                                             | Gemma-2B-it | ✅ Gemma-2-2B-it |
| -------------------------------------------------- | ----------- | --------------- |
| ✅ Newer architecture                               | ❌           | ✅               |
| ✅ Stronger instruction-following                   | ⚠️ Basic    | ✅ Advanced      |
| ✅ Improved reasoning ability                       | ⚠️ Weaker   | ✅ Strong        |
| ✅ Benchmarked to outperform GPT-3.5                | ❌           | ✅               |
| ✅ Supports chat-like prompts with chain-of-thought | ⚠️          | ✅               |
| ✅ Transparency tools (GemmaScope, ShieldGemma)     | ❌           | ✅               |
| ✅ LLM safety alignment for fairness/bias           | ❌           | ✅               |
